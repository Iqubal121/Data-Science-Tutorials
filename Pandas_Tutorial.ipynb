### **Pandas Series**

A pandas series is a one-dimensional labeled array capabel of holding data of any type(integer, string, float, Python Objects, etc.)

The axis labels are collectively called indexes.

The Pandas Series is noting but a column in Excel sheet.

Labels need not be unique but must be of a hashable type.



#### **Creating a Series**

Pandas Series is created by loading the datsets from existing storage

It can be a SQL database, a CSV file, or an Excel file.

Pandas Series can be created from lists, dictionaries, scalar values, etc.

image.png

```python
import numpy as np
import pandas as pd

# creating empty series
s = pd.Series()
print(s)

# creating the series
arr = np.array(['a', 'b', 'c', 'd', 'e'])
s = pd.Series(arr)
print(s)
```

```python
# How to create a series using dictionary

import pandas as pd

# create a dict
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

# create a series
s = pd.Series(dict1)
print(s)
```

```python
# add column using assign

import pandas as pd
import numpy as np

# read the dataset
data = pd.read_csv("sp500_companies.csv")

# define some city name for randomly generate
city_names = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "San Antonio", "San Diego", "Dallas", "San Jose"]

# Generate random city names for each row
random_cities = np.random.choice(city_names, size=len(data))

# Add the new 'City' column using assign
data = data.assign(City=random_cities)

# View the updated DataFrame
data.head()
```

```python
# add new column using list

# define some city name for randomly generate
types = ["Mini", "Macro", "Fortune", "Private", "Public", "Joint Venture", "Partnersgip", "Sole Prop"]
# Generate random city names for each row
random_cities = np.random.choice(types, size=len(data))

# Add the new 'City' column using assign
data["CompanyTypes"] = random_cities

# View the updated DataFrame
data.head()
```

```python
# we can add the columns using df.loc as well

status = ["Active", "Hold", "Inactive", "Released", "Concerned"]

# make the choice for random element
random_status = np.random.choice(status, size=len(data))

# add the new column using df.loc
data.loc[:, "Status"] = random_status

data
```

#### **Applying fucntions to group**

After splitting a data into a group, we apply a function to each group in order to do that we perform some operation they are.

1. *Aggregation:* --> We compute a summary statistic about each group.
2. *Transformation:* --> We perform some group-specific computations and return a like-indexed.

3. *Filtration:* --> We discard some groups, according to a group-wise computation that evalutes true and False.



**Aggregation**

--> It is a process in which we compute a summary statistic about each group.

--> It returns a single aggregated value for each group.

--> After splitting a data into groups unsibg groupby function, several aggregation operations can be performed on the grouped data.


```python
import pandas as pd

# read the dataset
data = pd.read_csv("/content/sp500_companies.csv")

# read top 5 data
data.head()
```

```python
grouped_data = data.groupby("Sector")
grouped_data.aggregate(np.sum)
```

```python
grouped_data = data.groupby(["Exchange","Sector"])
grouped_data.aggregate(np.sum)
```

```python
# we acn apply multiple function in the aggregate function as well
grouped_data = data.groupby(["Exchange","Sector"])
grouped_data["Marketcap"].agg([np.sum, np.mean, np.std])
```

```python
# we can apply different function on the dataframe columns as well
grouped_data = data.groupby(["Exchange","Sector"])
grouped_data.agg({"Marketcap": "min", "Currentprice":"sum"})
```

#### **Transformation**

--> Transformation is a process in which we perform some group-specific computations and return a like-indexed.

--> This methods return an object that is indexed the same as the one being grouped.

*Notes:*

1. Returns a result that is either the same size as the group chunk
2. Operate column-by-column on the group chunk.
3. Not perform in-place operations on the group chunk.


```python
import pandas as pd

# read the dataset
data = pd.read_csv("/content/sp500_companies.csv")

# read top 5 data
data.head(2)
```

```python
# using transformation on the data
grouped_data = data.groupby(["Exchange","Sector"])
grouped_data["Marketcap"].transform(lambda x: x.min())
```

#### **Filteration**

It is a process in which we discard some groups, according to a group wise computation that evaluates true or false.

We use filter method and apply some condition by which we filter group.



```python
import pandas as pd

# read the dataset
data = pd.read_csv("/content/sp500_companies.csv")

# read top 5 data
data.head()
```

```python
# filter the data
grouped_data = data.groupby(["Exchange"])
grouped_data.filter(lambda x: x["Currentprice"].mean() < 122)
```

#### **Combine multiple columns**

```python
# Add a new columns to an existing dataframe using datafrme.insert()

# import the libraries
import pandas as pd
import numpy as np

# import the dataset
data = pd.read_csv("sp500_companies.csv")

# Generate random values (for example, using np.random.rand for random float numbers between 0 and 1)
random_values = np.random.randint(50,3000, len(data))

# Insert the new column 'RandomValues' at position 2 (you can change this position)
data.insert(2, "Contract_Emp", random_values)

# View the updated DataFrame
data.head()
```

```python
# filter the dataset
total_emp = data[['Contract_Emp', "Fulltimeemployees"]]
total_emp.head()
```

```python
# set the index as Industry

# add the values of both and colmuns and store into one new columns using groupby
total_emp["Total_Emp"] = total_emp.groupby(data["Industry"])[["Contract_Emp", "Fulltimeemployees"]].transform('sum').sum(axis = 1)
total_emp.head()
```
